```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Sunspots.csv")
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df.drop(index=df[df["Monthly Mean Total Sunspot Number"] == 0].index,inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
df = df[~df['Year'].between(1790, 1830)]
df = df.reset_index(drop=True)
```

```python
df['Month_sin'] = np.sin(2 * np.pi * df['Month'] / 12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month'] / 12)
df['Year_sin'] = np.sin(2 * np.pi * (df['Year'] - df['Year'].min()) / 11)
df['Year_cos'] = np.cos(2 * np.pi * (df['Year'] - df['Year'].min()) / 11)
df['Sunspot_lag1'] = df['Monthly Mean Total Sunspot Number'].shift(1)
df['Sunspot_Lag_12'] = df['Monthly Mean Total Sunspot Number'].shift(12)
df['Diff_11Year'] = df['Monthly Mean Total Sunspot Number'].diff(periods=132)

df.dropna(inplace=True)
```

```python

df.to_csv("ForecastSunspot.csv", index=False)
```
